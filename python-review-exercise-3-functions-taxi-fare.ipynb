# Create a function called 'get_fare' that calculates the fare with given parameters

```python
# This code block is a function that 
# calculates the fare with two (2) given parameters
def get_fare(distance_traveled, 
             travel_time):
    
    # store constants in a variable
    base_fare = 40.0
    per_kilometer = 13.50
    per_minute = 2.0
    
    # calculates the distance traveled and travel time by multiplying to constants
    distance_traveled = distance_traveled * per_kilometer
    travel_time = travel_time * per_minute
    
    # calculates the total fare by adding values together
    total_fare = base_fare + distance_traveled + travel_time
    
    # returns variable total_fare
    return total_fare
```

# Create a function called 'main' that demonstrates the function 'get_fare'

```python
# This code block is a function that 
# demonstrates the function 'get_fare'
def main():
    
    # asks the user for input
    distance = int(input("Please Enter the Distance Traveled (km): "))
    time = int(input("Please Enter the Travel Time (mins): "))

    # calculates the total_fare using the get_fare function
    # by passing 2 parameters distance and time
    total_fare = int(get_fare(distance, time))
    
    # prints the calculated total_fare
    print(f"Total fare is: {total_fare}")
```

# Running and Debugging the Program

```python
# This code block
# runs the code for debugging
main()
```
